### Movies vs Tv shows

### Top 10 countries with most contents



```python
# Creating bar plot

top_countries = df['country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis', hue= top_countries.index)
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

### Organising Genres

```python
# Split listedin colomns into multiple genres
from collections import Counter

genre_list = df['listed_in'].dropna().str.split(', ')
genre_flat = [genre for sublist in genre_list for genre in sublist]
top_genres = pd.Series(Counter(genre_flat)).sort_values(ascending=False).head(10)

# Plot
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#C2C2F0',
          '#FFB3E6', '#FF6666', '#FFCC66', '#66CCCC', '#669999']

top_genres.plot(kind='barh', color=colors)

plt.title("Top 10 Genres on Netflix")
plt.xlabel("Number of Titles")
plt.gca().invert_yaxis()
plt.show()
```

### Visualise how many contents produced per years

```python
# Convert date_added to date and Time

df['date_added']= pd.to_datetime(df['date_added'], format='mixed', errors='coerce')

# Creating new coloum for years

df['year_added']= df['date_added'].dt.year

# Visualite the graph

df['year_added'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title("Content Added to Netflix Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

### Sorting popular genres by months


```python
# Extract month from date_added
df['month_added'] = pd.to_datetime(df['date_added']).dt.month_name()

# Explode the 'listed_in' column to separate multiple genres
df['genre'] = df['listed_in'].str.split(', ')
df_exploded = df.explode('genre')

# Count genres per month
genre_by_month = df_exploded.groupby(['month_added', 'genre']).size().reset_index(name='count')

# Sort within each month by count descending
genre_by_month_sorted = genre_by_month.sort_values(['month_added', 'count'], ascending=[True, False])

# Display result
print(genre_by_month_sorted.head(20))  # top 20 for a glance
```

#### Creating bar chart to visualise  genre by months

```python
# Clean dataset: drop rows with missing date_added
df_clean = df.dropna(subset=['date_added'])

# Extract month from date_added
df_clean['month_added'] = pd.to_datetime(df_clean['date_added']).dt.month_name()

# Explode 'listed_in' column into multiple rows for each genre
df_clean['genre'] = df_clean['listed_in'].str.split(', ')
df_exploded = df_clean.explode('genre')

# Count genres per month
genre_by_month = df_exploded.groupby(['month_added', 'genre']).size().reset_index(name='count')

# Sort month names in calendar order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
genre_by_month['month_added'] = pd.Categorical(genre_by_month['month_added'], categories=month_order, ordered=True)

# Sort by month and count
genre_by_month_sorted = genre_by_month.sort_values(['month_added', 'count'], ascending=[True, False])

# Plot: barplot with color hue by genre
plt.figure(figsize=(16, 8))
sns.barplot(data=genre_by_month_sorted, x='month_added', y='count', hue='genre', palette='tab20')

# Chart aesthetics
plt.title('Netflix Genres Added by Month (2021)', fontsize=18)
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Top genres by months

```python
# Find the top genre per month by selecting the genre with the highest count in each month
top_genre_by_month = genre_by_month_sorted.loc[genre_by_month_sorted.groupby('month_added')['count'].idxmax()]

# Plot a simpler barplot: top genre per month
plt.figure(figsize=(14, 6))
sns.barplot(data=top_genre_by_month, x='month_added', y='count', hue='genre', dodge=False, palette='tab20')

# Chart aesthetics
plt.title('Top Netflix Genre Added by Month (2021)', fontsize=18)
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.legend(title='Top Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Dram and International movies are overall pupular genre by months.
